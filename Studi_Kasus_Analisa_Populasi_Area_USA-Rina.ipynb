3. Import dan lakukan langkah awal Exploratory Data Analysis dengan melihat summary statistics yang ada pada dataframe. # Gunakan method head, tail, info, describe, dan value_count().
Note: Untuk memahami data lebih lanjut, lakukan analisa categorical variable terhadap column yang mempunyai value category (bukan angka/nilai) untuk mengetahui categori apa saja dan berapa jumlahnya dengan menggunakan fungsi .value_counts() yang sudah pernah dipelajari.

```python
#download dataset 
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
pop_side = pop.style.set_table_attributes("style = 'display:inline'"). set_caption('pop')
areas_side = areas.style.set_table_attributes("style = 'display:inline'"). set_caption('areas')
abbrevs_side = abbrevs.style.set_table_attributes("style = 'display:inline'"). set_caption('abbrevs')
display_html(pop_side._repr_html_() +' '+ areas_side._repr_html_()+' '+abbrevs_side._repr_html_(), raw=True)
```

Bisa kita lihat data population mempunyai column ``state/region``, ``ages``, ``year``, dan ``population``. Namun kita juga memerlukan data full nama kota yang kebetulan hanya di temukan pada dataset areas dan abbrevs. Pilih dataset abbrevs karena dia mempunyai column ``abbreviation`` yang bisa di merge dengan ``state/region`` pada dataset population. 

4. Kemudian lakukan merge() pada dataset pop dan abbrevs dengan tipe ``outer`` supaya tidak ada data yang hilang ketika terjadi mismatch, dan relasikan ``state/region`` terhadap ``abbrevation``.

5. Sekarang kita mempunyai dua column yang duplicate dan bisa di drop column ``abbreviation``. 

```python
abb_pop = pd.merge(abbrevs, pop, how='outer', left_on='abbreviation', right_on='state/region', left_index=False, right_index=False).drop('abbreviation', axis=1)
abb_pop.head()
```

6. Cek Jika ada Null/NaN value pada data dan analisa data mana yang mempunyai nilai null

```python
abb_pop.isna().any()
```

Bisa kita lihat column population dan state mempunyai NaN value yang berarti beberapa state/region tidak mempunyai match terhadap abbreviation yang kita delete sebelumnya

7. Check data mana saja yang mempunyai population data NaN value (.isnull())

```python
abb_pop[abb_pop['state'].isnull()]
```

Bisa kita lihat data ``state/region`` == ``PR`` / Puerto Rico sebelum tahun 2000 terdapat mismatch. Ini bisa di sebabkan missing data yang terdapat pada original dataset.

Dan yang terpenting adalah data ``state`` juga mempunyai beberapa missing values. Lihat data mana saja yang mempunyai nilai kosong pada column ``state``. 

8. Gunakan .unique() untuk mempermudah melihat data yang mempunyai NaN value.

```python
# hint: ____.loc[_____['____'].isnull(), 'state/region'].unique()
abb_pop.loc[abb_pop['state'].isnull(), 'state/region'].unique()
```

9. Kita bisa lihat hasil diatas jadi singkatan 'PR' dan 'USA' mempunyai mismatch, maka kita bisa hard-code atau secara manual assign data yang mempunyai singkatan 'PR' dan 'USA' dengan value 'Puerto Rico' dan 'United States'.

```python
# Hasilnya akan terlihat seperti dibawah, hasil akhir dataset kita akan mempunyai column yang lengkap (+ state & area)
```

```python
abb_pop.loc[abb_pop['state/region'] == 'PR', 'state'] = 'Puerto Rico'
abb_pop.loc[abb_pop['state/region'] == 'USA', 'state'] = 'United States'
# Check apakah sekarang column 'state' mempunyai nilai kosong atau tidak
abb_pop.isnull().any()
```

```python
abb_pop2 = abb_pop.fillna(abb_pop["population"].median())
abb_pop2.isnull().any()
```

Bisa kita perhatikan, sekarang column state tidak mempunyai nilai kosong dan sekarang dataset sudah siap untuk di merge kembali dengan dataset ``areas``. 

10. Buat variable baru dan gunakan fungsi merge pada column state untuk kedua dataset 
    - 1.) dataset yang sudah kita olah dan 2.) dataset areas

```python
abb_pop_area = pd.merge(abb_pop2, areas, how='outer')
abb_pop_area
```

11. Check kembali apakah ada hasil yang mismatch, jika masih ada sekarang kita bisa gunakan fungsi .dropna() untuk menghilangkan NaN values pada keseluruhan dataset

```python
# gunakan fungsi .isnull() dan .any() untuk mengetahui apakah masih ada NaN values pada data
abb_pop_area.isnull().any()
```

```python
abb_pop_area.dropna(inplace=True)
abb_pop_area.head()
```

```python
abb_pop_area.isnull().any()
```
